```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

data = pd.read_pickle("./data/2022/merged-data-2022.pkl")
X = data.drop('grav', axis=1)
y = data["grav"]

fig = plt.gcf()
plt.figure(figsize=(35,35))
sns.heatmap(X.corr(), annot=True)

plt.savefig("image/CorrelationMatrix_fullData.png")
plt.show()
```

```python
def compute_accuracy(X_ori, y_ori, random_state: int = 42, test_size: float = 0.3, classifier = RandomForestClassifier):
    """
    Calcul l'accuracy des données fournies en entrées en utilisant les paramètres initialisés

    :param data: Les données dont on veut calculer l'accuracy
    :return: L'accuracy comprise entre 0 et 1 sous forme de float
    """
    # Division des données
    X_train, X_test, y_train, y_test = train_test_split(
        X_ori,
        y_ori,
        test_size=test_size,
        random_state=random_state
    )
    # Modèle Défini en paramètre du constructeur
    classifier = classifier(random_state=random_state)
    classifier.fit(X_train, y_train)
    # Prédictions sur l'ensemble de test
    y_pred = classifier.predict(X_test)
    # Calcul de l'accuracy
    return accuracy_score(y_test, y_pred)
```

```python
from PcaOnData import PcaOnData

data = pd.read_pickle("./data/2022/merged-data-2022.pkl")
X = data.drop('grav', axis=1)
y = data["grav"]

acc_ori = compute_accuracy(X, y)
acc_without_catu = compute_accuracy(X.drop('catu', axis=1), y)
acc_without_locp = compute_accuracy(X.drop('locp', axis=1), y)
acc_without_Num_Acc = compute_accuracy(X.drop('Num_Acc', axis=1), y)
acc_without_com = compute_accuracy(X.drop('com', axis=1), y)
acc_without_etatp = compute_accuracy(X.drop('etatp', axis=1), y)
acc_without_actp = compute_accuracy(X.drop('actp', axis=1), y)
```

```python
print(f"acc_ori             = {acc_ori * 100:.4f}")
print(f"acc_without_catu    = {acc_without_catu * 100:.4f}")
print(f"acc_without_locp    = {acc_without_locp * 100:.4f}")
print(f"acc_without_Num_Acc = {acc_without_Num_Acc * 100:.4f}")
print(f"acc_without_com     = {acc_without_com * 100:.4f}")
print(f"acc_without_etatp   = {acc_without_etatp * 100:.4f}")
print(f"acc_without_actp    = {acc_without_actp * 100:.4f}")
```
